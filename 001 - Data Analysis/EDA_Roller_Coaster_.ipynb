# Steps for Understanding Data

Here are the top steps for understanding a dataset:

1. **Load the Data**: Import the dataset into your analysis environment.
2. **Inspect the Data Structure**: View the first few rows, check the shape, and list the columns.
3. **Check Data Types**: Identify the types of each column (numeric, categorical, etc.).
4. **Handle Missing Values**: Find and address missing or null values.
5. **Summary Statistics**: Calculate mean, median, min, max, and standard deviation for numeric columns.
6. **Visualize Distributions**: Plot histograms, boxplots, and other charts to see how data is distributed.
7. **Explore Relationships**: Use scatter plots, pairplots, and correlation matrices to find relationships between features.
8. **Analyze Categorical Features**: Count and visualize categories, such as types or manufacturers.
9. **Geographical Analysis**: If available, plot locations using latitude and longitude.
10. **Data Cleaning**: Remove duplicates, fix errors, and convert data types as needed.

Following these steps helps you gain a deep understanding of your data before further analysis or modeling.

STEP 3 - Feature Understanding

```python
df['Year_Introduced'].value_counts()
```

```python
# Visualize the top 10 years with most roller coaster introductions
df['Year_Introduced'].value_counts() \
    .head(10) \
    .plot(kind='bar', title='Top 10 Years with Most Roller Coaster Introductions', figsize=(10,6))
plt.xlabel('Year Introduced')
plt.ylabel('Number of Roller Coasters Introduced')
plt.show()
```

```python
# Visualize the distribution of roller coaster speeds
df['Speed_MPH'].dropna().plot(kind='hist', bins=20, title='Distribution of Roller Coaster Speeds', figsize=(10,6))
plt.xlabel('Speed (MPH)')  
plt.show()
```

```python
#density plot
sns.kdeplot(df['Speed_MPH'].dropna(), fill=True)
```

STEP 4: Feature Relationship

```python
#scatter plot to show relationship between height and speed
df.plot(kind='scatter', x='Height_FT', y='Speed_MPH', \
        title="Coaster Speed vs. Height", alpha=0.6, figsize=(10,6))
plt.show()
```

```python
#seaborn scatter plot
sns.scatterplot(data=df, x='Height_FT', y='Speed_MPH', hue='Year_Introduced', alpha=0.6)
plt.title("Coaster Speed vs. Height")
plt.show()
```

```python
#seaborn pairplot
sns.pairplot(df, vars=['Year_Introduced', 'Height_FT', 'Speed_MPH', 'Inversions_Clean', 'GForce_Clean'],
             hue='Type_Main')
plt.show()
```

```python
#correlation matrix
df[['Year_Introduced', 'Height_FT', 'Speed_MPH', 'Inversions_Clean', 'GForce_Clean']].dropna().corr()
```

```python
#heatmap of correlation matrix
df_corr = df[['Year_Introduced', 'Height_FT', 'Speed_MPH', 'Inversions_Clean', 'GForce_Clean']].dropna().corr()
plt.figure(figsize=(8,6))

sns.heatmap(df_corr, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Matrix of Roller Coaster Features')
```

STEP 5: What is the question we want to answer.
What are the locations with the fastest Roller Coasters?

```python
#What are the locations with the fastest Roller Coasters?
df['Location'].value_counts()

#query where location does not equal "Other"
df.query('Location != "Other"') \
    .groupby('Location')['Speed_MPH'] \
    .agg(['mean', 'count']) \
    .query('count >= 10') \
    .sort_values(by='mean', ascending=False) \
    .plot(kind='bar', title='Average Speed of Roller Coasters by Location (Min 10 Coasters)', figsize=(12,6))
plt.xlabel('Average Speed (MPH)')
plt.show()
```
